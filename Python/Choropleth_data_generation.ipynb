```python
import geopandas as gpd

lga = gpd.read_file("municipal-boundary/municipal-boundary.shp") # Load data

suburbs = gpd.read_file("GDA2020/vic_localities.shp").to_crs('4326') # convert crs 

melb_lga = lga[lga['name'].str.contains("Melbourne", case = False)] # filter city of Melbourne polygon

suburbs_in_melb = gpd.overlay(suburbs, melb_lga, how = 'intersection') # filter

suburbs_in_melb.to_file("city_of_melbourne_suburbs.geojson", driver = "GeoJSON") # Saving the suburbs 
```

```python
centroids = suburbs_in_melb.copy() # Copy the Melbourne suburbs layer

centroids = centroids.to_crs(epsg = 7855) # reproject 

centroids["geometry"] = centroids.geometry.centroid

centroids = centroids.to_crs(epsg = 4326)

centroids["centroid longitude"] = centroids.geometry.x

centroids["centroid latitude"]  = centroids.geometry.y

centroids.to_file("city_of_melbourne_suburbs_centroids.geojson", driver="GeoJSON") # Saving the suburbs (with centroid coordinates)
```
